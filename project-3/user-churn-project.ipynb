# User Churn Prediction with Machine Learning

Variables:
- ID – a sequential numbered index;
- label – binary target variable ('retained' vs 'churned') for if a user has churned anytime during the course of the month;
- sessions – the number of occurrence of a user opening the app during the month;
- drives – an occurrence of driving at least 1 km during the month;
- device – the type of device a user starts a session with;
- total_sessions – a model estimate of the total number of sessions since a user has onboarded;
- n_days_after_onboarding – the number of days since a user signed up for the app;
- total_navigations_fav1 – total navigations since onboarding to the user’s favorite place 1;
- total_navigations_fav2 – total navigations since onboarding to the user’s favorite place 2;
- driven_km_drives – total kilometers driven during the month;
- duration_minutes_drives – total duration driven in minutes during the month;
- activity_days – number of days the user opens the app during the month;
- driving_days – number of days the user drives (at least 1 km) during the month.

Objectives: 
- analyse user records to understand which users tend to churn;
- build a machine learning model to predict if a user will churn.

## 1. Data Preparation and Initial Exploratory Data Analysis

```python
# Import libraries

# For data manipulation, analysis, and visualisation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import randint

# For machine learning
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from xgboost import XGBClassifier
```

```python
# Since there are 700 missing values in the dataset, I will use predictive modelling to impute the missing values
# Split the data into rows with non-missing and missing values
df_non_missing = df[df['label'].notna()]  # Rows with no missing values
df_missing = df[df['label'].isna()]  # Rows with missing values
```

```python
# Create a copy of df_non_missing to avoid modifying a view
df_non_missing = df_non_missing.copy()

# Encode the 'device' and 'label' columns
df_non_missing.loc[:, 'device_encoded'] = df_non_missing['device'].map({'Android': 0, 'iPhone': 1})
df_non_missing.loc[:, 'label_encoded'] = df_non_missing['label'].map({'retained': 0, 'churned': 1})
```

```python
# Create a copy of df_missing to avoid modifying a view
df_missing = df_missing.copy()

# Encode the 'device' column
df_missing.loc[:, 'device_encoded'] = df_missing['device'].map({'Android': 0, 'iPhone': 1})
```

```python
# Separate features and target variable
X_train = df_non_missing.drop(['label', 'label_encoded', 'device'], axis=1)
y_train = df_non_missing['label_encoded']

# Split the data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Initialise the Random Forest model
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
# Prepare data for the rows with missing values
X_missing = df_missing.drop(['label', 'device'], axis=1)

# Make predictions
predicted_values = model.predict(X_missing)

# Assign predicted values
df_missing['label_encoded'] = predicted_values
```

```python
# Combine the train_df and missing_df back into the original df
df_new = pd.concat([df_non_missing, df_missing], axis=0)

# Sort indices
df_new = df_new.sort_index()

# Map the encoded values back to the original 'label' column
df_new['label'] = df_new['label_encoded'].map({0: 'retained', 1: 'churned'})
```

### 1.3. Outliers

```python
# Write a function to detect outliers
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25) # Calculate 25th percentile
    Q3 = df[column].quantile(0.75) # Calculate 75th percentile
    IQR = Q3 - Q1 # Calculate IQR (Interquartile Range)
    lower_bound = Q1 - 1.5 * IQR # Define lower bound
    upper_bound = Q3 + 1.5 * IQR # Define upper bound
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)] # Find outliers
    return outliers

# Detect outliers
outliers = detect_outliers(df_new, ['sessions', 'drives', 'total_sessions', 'n_days_after_onboarding', 'total_navigations_fav1', 
                                    'total_navigations_fav2', 'driven_km_drives', 'duration_minutes_drives', 'activity_days', 'driving_days'])
print(outliers.count())
```

```python
# Write a function to visualise outliers
def visualise_outliers(df, columns):
    num_rows = (len(columns) + 1) // 2 # Ensure that the last row will display a plot even if the number of plots is odd
    fig, axes = plt.subplots(num_rows, 2, figsize=(15, 5*num_rows)) # Plot displaying 2 plots on each row
    axes = axes.flatten() # Convert the axes array (which is a 2D array when there are multiple rows and columns of subplots) into a 1D array

    # If there is only 1 plot, then I need to convert it into a list so that I can use it in the same way as when there are multiple plots
    if len(columns) == 1:
        axes = [axes]

    # Plot boxplots
    for i, column in enumerate(columns):
        sns.boxplot(x=df[column], ax=axes[i])
        axes[i].set_title(f"Distribution of '{column}' with Outliers")
        axes[i].set_xlabel('')

        # Turn off unused axes to avoid displaying empty plots
        for j in range (len(columns), len(axes)):
            axes[j].axis('off')
    
    plt.tight_layout(pad=2)
    plt.show()

# Visualise outliers
visualise_outliers(df_new, ['sessions', 'drives', 'total_sessions', 'total_navigations_fav1', 'total_navigations_fav2', 
                         'driven_km_drives', 'duration_minutes_drives'])
```

#### Comments
I will set up an A/B test to test whether churn depends on the device type.

Hypotheses:
- Null Hypothesis (H0): The churn rate is the same for users on both devices (iPhone and Android).
- Alternative Hypothesis (H1): The churn rate differs between users on iPhone and users on Android.

Groups:
- Group A: Users who have an iPhone.
- Group B: Users who have an Android.

```python
# Calculate churn rates for iPhone and Android users
iphone_churn_rate = df_new[df_new['device_encoded'] == 1]['label_encoded'].mean()  # Churn rate for iPhone users
android_churn_rate = df_new[df_new['device_encoded'] == 0]['label_encoded'].mean()  # Churn rate for Android users

print(f"Churn rate for iPhone users: {iphone_churn_rate:.2f}")
print(f"Churn rate for Android users: {android_churn_rate:.2f}")
```

```python
# Create two groups: number of churned users and total users per device type
churned_iphone = df_new[(df_new['device_encoded'] == 1) & (df_new['label_encoded'] == 1)].shape[0]
total_iphone = df_new[df_new['device_encoded'] == 1].shape[0]

churned_android = df_new[(df_new['device_encoded'] == 0) & (df_new['label_encoded'] == 1)].shape[0]
total_android = df_new[df_new['device_encoded'] == 0].shape[0]

# Perform the Z-test for proportions
counts = [churned_iphone, churned_android]
nobs = [total_iphone, total_android]
stat, p_value = proportions_ztest(counts, nobs)

# Check the p-value
print(f"Z-test statistic: {stat}")
print(f"P-value: {p_value}")

# If p-value is below a significance level (e.g., 0.05), reject the null hypothesis
if p_value < 0.05:
    print("We reject the null hypothesis: Churn rates differ by device.")
else:
    print("We fail to reject the null hypothesis: Churn rates do not differ by device.")
```

#### Comments
- Z-test showed that churn rates do not differ by device.

## 3. Predicting Churn Using Machine Learning

```python
# Select features and the target value
features = ['sessions', 'drives', 'total_sessions', 'n_days_after_onboarding', 'total_navigations_fav1', 'total_navigations_fav2', 
            'driven_km_drives', 'duration_minutes_drives', 'activity_days', 'driving_days', 'device_encoded', 'session_engagement_ratio', 
            'avg_drive_distance', 'drives_per_day', 'total_navigations', 'navigation_per_session', 'avg_drive_duration']
X = df_new[features]
y = df_new['label_encoded']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### 3.1. Logistic Regression Model

```python
# Initialise and train the Logistic Regression model
lr_model = LogisticRegression(
    class_weight='balanced',  # Helps the model pay more attention to the minority class
    C=0.01,                   # Smaller C = stronger regularisation = less overfitting
    penalty='l2',             # Default, helps control complexity
    solver='liblinear',       # Good for smaller datasets
    random_state=42,
    max_iter=5000            
)
lr_model.fit(X_train, y_train)

# Make predictions
lr_y_pred = lr_model.predict(X_test)
lr_y_prob = lr_model.predict_proba(X_test)[:, 1]

# Evaluate the model
print("Classification Report (Logistic Regression Model):\n", classification_report(y_test, lr_y_pred))
print("Confusion Matrix (Logistic Regression Model):\n", confusion_matrix(y_test, lr_y_pred))
print("ROC AUC Score (Logistic Regression Model):", roc_auc_score(y_test, lr_y_prob))
```

```python
# Plot feature importance

# Get feature names and coefficients
lr_coefficients = lr_model.coef_[0]
lr_feature_importance = pd.Series(lr_coefficients, index=features)

# Sort by absolute value (most impactful)
lr_feature_importance = lr_feature_importance.reindex(lr_feature_importance.abs().sort_values(ascending=False).index)

# Plot
plt.figure(figsize=(10, 6))
lr_feature_importance.plot(kind='barh')
plt.title('Feature Importance (Logistic Regression Model)')
plt.xlabel('')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Perform hyperparameter tuning with GridSearchCV

# Define the parameter grid for tuning
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],           # Regularisation strength
    'penalty': ['l1', 'l2'],                # Regularisation type
    'solver': ['liblinear', 'saga'],        # Solvers that work with l1 and l2 penalties
    'max_iter': [1000, 5000],               # Number of iterations to ensure convergence
    'class_weight': ['balanced', None]      # Handle class imbalance
}

# Set up GridSearchCV with cross-validation
grid_search = GridSearchCV(estimator=lr_model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

# Fit the model on training data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_lr_params = grid_search.best_params_

# Get and fit the best model
best_lr_model = grid_search.best_estimator_
best_lr_model.fit(X_train, y_train)

# Make predictions with the best model
best_lr_y_pred = best_lr_model.predict(X_test)
best_lr_y_prob = best_lr_model.predict_proba(X_test)[:, 1]

# Print best hyperparameters and evaluate the model
print("Best Parameters (Tuned Logistic Regression Model):\n", best_lr_params)
print("Classification Report (Tuned Logistic Regression Model):\n", classification_report(y_test, best_lr_y_pred))
print("Confusion Matrix (Tuned Logistic Regression Model):\n", confusion_matrix(y_test, best_lr_y_pred))
print("ROC AUC Score (Tuned Logistic Regression Model):", roc_auc_score(y_test, best_lr_y_prob))
```

### 3.2. Random Forest Model

```python
# Initialise and train the Random Forest model
rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
rf_y_pred = rf_model.predict(X_test)
rf_y_prob = rf_model.predict_proba(X_test)[:, 1]

# Evaluate the model
print("Classification Report (Random Forest Model):\n", classification_report(y_test, rf_y_pred))
print("Confusion Matrix (Random Forest Model):\n", confusion_matrix(y_test, rf_y_pred))
print("ROC AUC Score (Random Forest Model):", roc_auc_score(y_test, rf_y_prob))
```

```python
# Perform hyperparameter tuning with RandomizedSearchCV

# Define the hyperparameter space for RandomizedSearchCV
param_dist = {
    'n_estimators': [100, 200],          # Use fewer trees (100 or 200)
    'max_depth': [None, 10, 20],          # Limit depth of trees to 10 and 20
    'min_samples_split': randint(2, 6),   # Smaller range for splitting
    'min_samples_leaf': randint(1, 4),    # Smaller range for leaf nodes
    'bootstrap': [True],                  # Use only bootstrap sampling
}

# Initialise RandomizedSearchCV with reduced iterations and cross-validation folds
random_search = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_dist,
    n_iter=5,     # Reduce iterations to 5
    cv=2,         # Use 2-fold cross-validation to speed up
    n_jobs=-1,    # Use all available CPU cores
    verbose=2,    # Print progress
    random_state=42
)

# Perform hyperparameter tuning with RandomizedSearchCV
random_search.fit(X_train, y_train)

# Get the best hyperparameters
best_rf_params = random_search.best_params_

# Get and fit the best model
best_rf_model = random_search.best_estimator_
best_rf_model.fit(X_train, y_train)

# Make predictions using the best model
best_rf_y_pred = best_rf_model.predict(X_test)
best_rf_y_prob = best_rf_model.predict_proba(X_test)[:, 1]

# Print best hyperparameters and evaluate the model
print("Best Hyperparameters (Tuned Random Forest Model):\n", best_rf_params)
print("Classification Report (Tuned Random Forest Model):\n", classification_report(y_test, best_rf_y_pred))
print("Confusion Matrix (Tuned Random Forest Model):\n", confusion_matrix(y_test, best_rf_y_pred))
print("ROC AUC Score (Tuned Random Forest Model):", roc_auc_score(y_test, best_rf_y_prob))
```
