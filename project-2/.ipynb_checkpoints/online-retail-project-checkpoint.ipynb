# Optimising Sales Strategies: Sales Trends and Product Performance Analysis and Prediction with Machine Learning

This is a transactional dataset which contains all the transactions occurring between 01/12/2010 and 09/12/2011 for a UK-based and registered non-store online retail. The company mainly sells unique all-occasion gifts. Many customers of the company are wholesalers.

Variables:
- InvoiceNo - a 6-digit integral number uniquely assigned to each transaction. If this code starts with letter 'c', it indicates a cancellation;
- StockCode - a 5-digit integral number uniquely assigned to each distinct product;
- Description - product name;
- Quantity - the quantities of each product (item) per transaction;
- InvoiceDate - the day and time when each transaction was generated;
- UnitPrice - product price per unit (sterling);
- CustomerID - a 5-digit integral number uniquely assigned to each customer;
- Country - the name of the country where each customer resides.

References:
- Chen, D. (2015). Online Retail [Dataset]. UCI Machine Learning Repository. https://doi.org/10.24432/C5BW33.

Objectives:
- Understand the trends and patterns in sales data.
- Analyse product performance, including best-sellers, underperforming products, and returns.
- Use machine learning to forecast future sales and predict returns.

## 1. Initial Analysis: Data Cleaning and Exploration

```python
# Import libraries

# For data manipulation
import pandas as pd
import numpy as np
from scipy import stats
import re

# For data visualisation
import matplotlib.pyplot as plt
import seaborn as sns

# For machine learning
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
```

```python
# Load the dataset
df = pd.read_csv('online-retail-data.csv')

# Display the first 5 rows
df.head()
```

### 1.1. Duplicates

```python
# Check for duplicates
df.duplicated().sum()

# Drop duplicates
df = df.drop_duplicates()
```

### 1.2. Missing Values and Data Types

```python
# Check for missing values
df.isnull().sum()
```

```python
# Replace '?' with None
df['Description'] = df['Description'].replace('?', None)

# Find missing Description values
missing_description = df[df['Description'].isna()]['StockCode'].unique()

# Fill missing Description values based on StockCode (using the mode of Description)
df['Description'] = df.groupby('StockCode')['Description'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x)
```

```python
# Again, check for missing values
print(df.isnull().sum())
```

```python
# Since there still are 112 missing Description values, I will fill them with 'Unknown' values
df['Description'] = df['Description'].fillna('Unknown')
```

```python
# Regular expression to exclude any description containing 'charge', 'fee', etc.
exclude_keywords = [r'\bfee\b', r'\bcharge\b', r'\bcharges\b', r'\bdiscount\b', r'\bpostage\b', r'\bshipping\b', r'\bhandling\b', r'\bmanual\b', 
                    r'\bcommission\b', r'\bsample\b', r'\bsamples\b']

# Create a pattern to match any of the exclude keywords
exclude_pattern = '|'.join(exclude_keywords)

# Create a new column 'IsCharge' where True represents a charge and False means it is a product
df['IsCharge'] = df['Description'].str.contains(exclude_pattern, case=False, na=False)
```

```python
# Check variable types
df.dtypes
```

```python
# Convert 'CustomerID' column type and fill the missing values with Unknown values
df['CustomerID'] = pd.to_numeric(df['CustomerID'], errors='coerce')
df['CustomerID'] = df['CustomerID'].fillna('Unknown')
df['CustomerID'] = df['CustomerID'].apply(lambda x: int(x) if isinstance(x, float) and not pd.isna(x) else x)
```

```python
# Convert 'InvoiceDate' column type
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%d/%m/%Y %H:%M')
```

```python
# Again, check for missing values
df.isnull().sum()
```

### 1.3. Outliers

```python
# Write a function to detect outliers

def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25) # Calculate 25th percentile
    Q3 = df[column].quantile(0.75) # Calculate 75th percentile
    IQR = Q3 - Q1 # Calculate IQR (Interquartile Range)
    lower_bound = Q1 - 1.5 * IQR # Define lower bound
    upper_bound = Q3 + 1.5 * IQR # Define upper bound
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)] # Find outliers
    
    # Return whether there are outliers
    return outliers

# Detect ouliers
outliers = detect_outliers(df, ['Quantity', 'UnitPrice'])
print(outliers.count())
```

```python
# Rows where Description is 'Adjust bad debt' seem to be errors
# Drop these rows
df = df[df['Description'] != 'Adjust bad debt']
```

```python
# Seperate cancellations from sales
df['IsCancellation'] = df['InvoiceNo'].str.startswith('C')
cancellations = df[df['IsCancellation'] == True]
print('Number of cancellations: {}.'.format(len(cancellations)))
```

```python
# Find the number of orders with negative Quantity values that were not cancelled
neg_q = df[(df['IsCancellation'] == False) & (df['Quantity'] < 0)]
print('Number of orders with negative Quantity values that were not cancelled: {}.'.format(len(neg_q)))
```

```python
# Investigate orders with negative Quantity values that were not cancelled
print(neg_q['CustomerID'].unique())
print(neg_q['UnitPrice'].unique())
```

```python
# Orders with negative Quantity values that were not cancelled have UnitPrice values of 0 and missing values of CustomerID
# Drop these rows
df = df[~((df['IsCancellation'] == False) & (df['Quantity'] < 0))]
```

```python
# Create a column to distinguish between individual buyers and wholesalers

# Define mean value of Quantity
mean_quantity = df['Quantity'].mean()

# Individuals: Quantity between negative and positive mean
# df['IsIndividualBuyer'] = df[(df['Quantity'] <= mean_quantity) & (df['Quantity'] >= -mean_quantity)]

# Wholesalers: Quantity greater than the positive mean or less than the negative mean
df['IsWholesaler'] = (df['Quantity'] > mean_quantity) | (df['Quantity'] < -mean_quantity)
```

```python
# Detect outliers
def detect_outliers(df, column, threshold=3.5):
    """Detects outliers using the Modified Z-Score method based on MAD"""
    median = df[column].median()
    mad = stats.median_abs_deviation(df[column])
    mod_z_scores = 0.6745 * (df[column] - median) / mad # Compute Modified Z-Scores
    outliers = df[abs(mod_z_scores) > threshold] # Filter outliers
    return outliers

# Detect outliers
outliers = {col: detect_outliers(df, col, threshold=3.5) for col in ['Quantity', 'UnitPrice']}

# Print the number of outliers per column
print("Outliers:")
for col, outliers in outliers.items():
    print(f" - {col}: {len(outliers)}")
```

```python
# Cap outliers
def cap_outliers(df, column, threshold=3.5):
    median = df[column].median()
    mad = stats.median_abs_deviation(df[column])
    
    # Compute Modified Z-Scores
    mod_z_scores = 0.6745 * (df[column] - median) / mad
    
    # Define upper and lower limits
    lower_bound = df[column][mod_z_scores > -threshold].min()
    upper_bound = df[column][mod_z_scores < threshold].max()

    # Cap values
    df.loc[df[column] < lower_bound, column] = lower_bound
    df.loc[df[column] > upper_bound, column] = upper_bound
    
    print(f"{column} | Lower bound: {lower_bound} | Upper bound: {upper_bound}")

# Apply to df
for column in ['Quantity', 'UnitPrice']:
    cap_outliers(df, column, threshold=3.5)
```

## 2. Exploratory Data Analysis and Feature Engineering: Advanced Analysis and Data Visualisation

```python
# Calculate TotalPrice
df.loc[:, 'TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Create a new df to store only products (not charges)
df_products = df[df['IsCharge'] == False]
```

```python
# Plot total earnings over time
df_products = df_products.copy()
df_products['YearMonth'] = df_products['InvoiceDate'].dt.to_period('M')
monthly_sales = df_products.groupby('YearMonth')['TotalPrice'].sum()
plt.figure(figsize=(12, 5))
monthly_sales.plot(kind='line', marker='o', color='purple')
plt.title('Total Earnings Over Time')
plt.xlabel('')
plt.ylabel('Total Earnings (£)')
plt.show()
```

#### Comments
- Until August, total earnings never exceeded £500,000.
- Since August, there had been a gradual increase with a peak of £900,000 in November.
- Since November, there has been a sharp decline in total earnings with the last data point of £300,000.

```python
# Find and plot top and worst 10 products by total earnings

# Get top 10 products by TotalPrice
top_10_products = df_products.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)

# Get worst 10 products by TotalPrice
worst_10_products = df_products.groupby('Description')['TotalPrice'].sum().sort_values(ascending=True).head(10)

# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot the top 10 products
sorted_idx_top = np.argsort(top_10_products)[::-1]
colors_top = sns.color_palette('Purples', len(sorted_idx_top) + 5)[5:]
colors_top = list(reversed(colors_top))
top_10_products.plot(kind='bar', color=colors_top, ax=ax[0])
ax[0].set_title('Top 10 Products by Total Earnings')
ax[0].set_xlabel('')
ax[0].set_ylabel('Total Earnings (£)')
ax[0].set_xticklabels(top_10_products.index, rotation=45, ha='right')

# Plot the worst 10 products
sorted_idx_worst = np.argsort(worst_10_products)
colors_worst = sns.color_palette('Purples', len(sorted_idx_worst) + 5)[5:]
colors_worst = list(reversed(colors_worst))
worst_10_products.plot(kind='bar', color=colors_worst, ax=ax[1])
ax[1].set_title('Worst 10 Products by Total Earnings')
ax[1].set_xlabel('')
ax[1].set_ylabel('Total Earnings (£)')
ax[1].set_xticklabels(worst_10_products.index, rotation=45, ha='right')

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

## 3. Predicting Cancellations Using Machine Learning

```python
df_products['Month'] = df_products['InvoiceDate'].dt.month
df_products['DayOfWeek'] = df_products['InvoiceDate'].dt.dayofweek
```

```python
# Convert 'IsWholesaler' column to integers (True = 1, False = 0)
df_products['IsWholesaler'] = df_products['IsWholesaler'].astype(int)

# Encode 'StockCode' using LabelEncoder
label_encoder = LabelEncoder()
df_products['StockCodeEncoded'] = label_encoder.fit_transform(df_products['StockCode'])

# Select features
features = ['StockCodeEncoded', 'Quantity', 'UnitPrice', 'Month', 'DayOfWeek', 'IsWholesaler', 'Country']
X = df_products[features]

# Apply one-hot encoding to 'Country' column
X = pd.get_dummies(X, columns=['Country'], drop_first=True)

# Select a target value
y = df_products['IsCancellation']

# Standardise the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform 70-20-10 split
X_train_full, X_test, y_train_full, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.125, random_state=42)

# Convert back to df with proper column names
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
X_val = pd.DataFrame(X_val, columns=X.columns)

print(f"Training set: {X_train.shape}")
print(f"Testing set: {X_test.shape}")
print(f"Validation set: {X_val.shape}")
print('\n')

# Train the Logistic Regression model
lr_model = LogisticRegression(
    class_weight='balanced',  # This helps with class imbalance
    C=0.01,
    penalty='l2',
    solver='liblinear',
    random_state=42
)
lr_model.fit(X_train, y_train)

# Predict on the test set
y_pred = lr_model.predict(X_test)

# Evaluate the Logistic Regression model
train_acc = accuracy_score(y_train, lr_model.predict(X_train))
test_acc = accuracy_score(y_test, y_pred)
val_acc = accuracy_score(y_val, lr_model.predict(X_val))

print("Accuracy (Logistic Regression Model):")
print(f"Training Accuracy: {train_acc:.4f}")
print(f"Testing Accuracy: {test_acc:.4f}")
print(f"Validation Accuracy: {val_acc:.4f}")
print('\n')
print('Accuracy (Logistic Regression Model): ', accuracy_score(y_test, y_pred))
print('\n')
print('Classification Report (Logistic Regression Model):')
print(classification_report(y_test, y_pred))
print('\n')
print('Confusion Matrix (Logistic Regression Model):')
print(confusion_matrix(y_test, y_pred))
print('\n')
cv_scores = cross_val_score(lr_model, X_scaled, y, cv=5)
print('Cross-Validation Scores (Logistic Regression Model): ', cv_scores)
print('\n')
print('Average Cross-Validation Score (Logistic Regression Model): ', cv_scores.mean())
```

#### Comments
- The model shows near-perfect results. It might suggest that the model is overfitting or a data imbalance issue (since there is only 1.6% of cancelled orders in the dataset). However, after adding *class_weight='balanced'*, the model no more misses *True* values - which is the most important point since there are very little *True* values in the dataset; hence, the data imbalance issue was fixed.
